```python
import gymnasium as gym
import matplotlib.pyplot as plt
from IPython.display import clear_output
import time
```

```python
# Create MountainCar environment
env = gym.make("MountainCar-v0", render_mode="rgb_array")
obs, _ = env.reset()

total_reward = 0
done = False

while not done:
    # Render current frame as image
    frame = env.render()

    # Clear previous output and display the frame
    clear_output(wait=True)
    plt.imshow(frame)
    plt.axis("off")
    plt.title("MountainCar-v0")
    plt.show()

    # Take random action
    action = env.action_space.sample()
    obs, reward, terminated, truncated, _ = env.step(action)
    done = terminated or truncated
    total_reward += reward
    time.sleep(0.03)  # control playback speed

env.close()
print("Total Reward:", total_reward)

```

**Environment 1: MountainCar-v0**

Observation Space:
- Type: Box(2,)
- Description: A 2D continuous vector representing:
  1. Position of the car along the x-axis (range: -1.2 to 0.6)
  2. Velocity of the car (range: -0.07 to 0.07)

Action Space:
- Type: Discrete(3)
- Actions:
  0 → Push Left
  1 → No Push
  2 → Push Right

Goal / Objective:
- Get the underpowered car to the top of the right hill (position ≥ 0.5) by building momentum through swinging.

Episode Length:
- Maximum of 200 steps per episode.

Challenges:
- The engine is too weak to drive up the hill directly.
- The agent must learn to reverse, build momentum, and then swing up the opposite hill.
- Requires learning of delayed rewards and planning ahead.

```python
import gymnasium as gym
import matplotlib.pyplot as plt
from IPython.display import clear_output
import time

# Use the latest version
env = gym.make("LunarLander-v3", render_mode="rgb_array")
obs, _ = env.reset(seed=42)

total_reward = 0
done = False

while not done:
    frame = env.render()
    clear_output(wait=True)
    plt.imshow(frame)
    plt.axis("off")
    plt.title("LunarLander-v3")
    plt.show()

    action = env.action_space.sample()  # random agent
    obs, reward, terminated, truncated, _ = env.step(action)
    done = terminated or truncated
    total_reward += reward
    time.sleep(0.03)

env.close()
print("Total Reward:", total_reward)
```

**Environment 2: LunarLander-v3**


Observation Space:
- Type: Box(8,)
- Description: An 8-dimensional vector representing:
  1. X position
  2. Y position
  3. X velocity
  4. Y velocity
  5. Angle of the lander
  6. Angular velocity
  7. Left leg contact (boolean)
  8. Right leg contact (boolean)

Action Space:
- Type: Discrete(4)
- Actions:
  0 → Do nothing
  1 → Fire left engine
  2 → Fire main (downward) engine
  3 → Fire right engine

Goal / Objective:
- Land softly on the central landing pad with minimal velocity and tilt, keeping both legs in contact.

Episode Length:
- Maximum of 1000 steps (usually ends on crash or successful landing).

Challenges:
- Physics-heavy simulation with gravity, inertia, and rotation.
- Requires balancing multiple forces and precise control of thrust.
- Penalizes hard landings, drifting, and crashes.
